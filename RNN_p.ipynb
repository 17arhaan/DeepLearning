```python
import torch
import matplotlib.pyplot as plt 
from torch import nn 
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, Dataset
```

```python
import matplotlib.pyplot as plt
# Sample data for Actual vs Predicted
actual = [1, 2, 3, 4, 5]
predicted = [1.1, 1.9, 3.2, 3.8, 5.1]
# Plot
plt.figure(figsize=(6, 4))
plt.plot(actual, 'b-', label='Actual')
plt.plot(predicted, 'r--', label='Predicted')
plt.title('Actual vs Predicted')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Sample data for Error vs Loss
steps = np.arange(0, 100, 0.01)
loss1 = 1 / (steps + 1)
# Plot
plt.figure(figsize=(6, 4))
plt.plot(steps, loss1, 'b-', label='Loss')
plt.xlabel('Error')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
